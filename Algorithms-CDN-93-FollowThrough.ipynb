## In-Place Algorithms

#### Syntax

```python
#modifying the original thing 

def test(a):
    print(f"Memory address for a:{(hex(id(a)))}")
    a.append(4)
    
b = [1, 2, 3]
print(b)
print(f"Memory address for b:{(hex(id(b)))}")
test(b)
print(b)
```

#### Out of Place Algorithm

```python
# creating a new space in memory to work with the data 

def test(a):
    a = a[:]
    print(f"Memory address for a:{(hex(id(a)))}")
    a.append(4)
    return a
    
b = [1, 2, 3]
print(b)
print(f"Memory address for b:{(hex(id(b)))}")
print(test(b))
print(f"Memory address for b:{(hex(id(test(b))))}")
print(b)
```

## Two Pointers
used to compare two items, usually iterables

#### Syntax

```python
#use this when you are moving through an iterable and comparing items within that iterable

def reverse_list(lst): 
    lst = lst[:]
    
    left = 0 
    right = len(lst) - 1
    while left < right: 
        lst[left], lst[right] = lst[right], lst[left]
        left += 1 
        right -= 1
    return lst
    
    
    
print(reverse_list([1, 2, 3, 4, 5, 6]))
    
```

#### Bubble Sort

Worst Case: O(n^2) Time - O(1) Space <-- Big O Notation

```python
import timeit 

hm = [11, 10, 5, 15, 100]

def bubble_sort(lst): 
    lst = lst[:]
    end_idx = len(lst) - 1
    #we don't need to define pointers 
    #variable that holds the state:
    is_sorted = False 
    while not is_sorted: #this flag only gets considered again when the entire while loop gets back.
        is_sorted = True
        for i in range(end_idx): 
            if lst[i] > lst[i+1]: 
                lst[i], lst[i+1] = lst[i+1], lst[i]
                is_sorted = False
        end_idx -= 1
    return lst

bubble_sort(hm)
        
```

##### Insertion Sort

Worst Case: O(n^2) time - O(1)space

```python
input_list = [11,10, 5 , 15, 100]

def insertion_sort(lst): 
    lst = lst[:]
    for i in range(1, len(lst)):
        j = i
        while j > 0 and lst[j] < lst[j - 1]: 
            lst[j], lst[j-1] = lst[j-1], lst[j]
            j-= 1
    return lst
        

print(insertion_sort(input_list))
```

## Merge Sort

#### How it Works

1. Split everything as evenly as possible into their own subarrays
2. move from left to right merging out groups together 
3. compare items and merge into correct position 
4. continue until the array is sorted

```python
input_list = [11,10, 5 , 15, 100]

def merge_sort(list_): 
    print("Splitting", list_)
    # Not a while loop because we are using recursion.
    if len(list_) > 1: 
        mid = len(list_) // 2
        left_half = list_[:mid]
        right_half = list_[mid:]
    #call the function again and it will go again
    # until the condition is met in line 6
        merge_sort(left_half)
        merge_sort(right_half)
        
        i = 0 #pointer for the left 
        j = 0 #pointer for the right
        k = 0 #pointer for the main array
        #step 2
        while i < len(left_half) and j < len(right_half): 
            if left_half[i] < right_half[j]: 
                list_[k] = left_half[i]
                i+= 1
            else: 
                list_[k] = right_half[j]
                j+= 1
            
            k+=1
       #step 3 
        while i < len(left_half): 
            list_[k] = left_half[i]
            i += 1 
            k += 1 
            
            
        while j < len(right_half): 
            list_[k] = right_half[j]
            j += 1 
            k += 1 
            
    return list_
merge_sort(input_list)
```

# Binary Search

The Binary Search algorithm works by finding the number in the middle of a given array and comparing it to the target. Given that the array is sorted

* The worst case run time for this algorithm is `O(log(n))`

```python
the_list = [1, 2, 3, 5, 6, 10, 11]

the_list.index(6)
```

The goal is to reimplement this without using build in methods

```python
def binary_search(lst, target):
    mid = len(lst)// 2
    
    if lst[mid] == target: 
        return mid
    elif lst[mid] > target: 
        return binary_search(lst[:mid], target) 
    else: 
        return mid + binary_search(lst[mid:], target) 

    
    
    
binary_search([1,5,8,9],9)
    
```

### Exercise #1 <br>
<p>Reverse the list below in-place using an in-place algorithm.<br>For extra credit: Reverse the strings at the same time.</p>

```python
words = ['this' , 'is', 'a', 'sentence', '.']

def reverser(lst): 
    left = 0 
    right = len(lst) - 1
    while left < right: 
        lst[left], lst[right] = lst[right], lst[left]
        left += 1 
        right -= 1
    extra = [] 
    for word in lst: 
        first = 0
        last = len(word) - 1
        while first < last:
            word = list(word)
            word[first], word[last] = word[last], word[first]
            first += 1 
            last -= 1 
        extra.append("".join(word))

    return "Switched words:", lst, "Switched letters", extra

print(reverser(words))

```

### Exercise #2 <br>
<p>Create a function that counts how many distinct words are in the string below, then outputs a dictionary with the words as the key and the value as the amount of times that word appears in the string.<br>Should output:<br>{'a': 5,<br>
 'abstract': 1,<br>
 'an': 3,<br>
 'array': 2, ... etc...</p>

```python
a_text = 'In computing, a hash table hash map is a data structure which implements an associative array abstract data type, a structure that can map keys to values. A hash table uses a hash function to compute an index into an array of buckets or slots from which the desired value can be found'

from collections import Counter 
from pprint import pprint

# --- with Counter ---

def distinction(string):
    special_char = '@_!#$%^&*()<>?/\|}{~:;[]'
    for i in special_char: 
        string = string.replace(i, "")
    str_lst = string.lower().split(" ")
    str_dict = Counter(str_lst)
    return str_dict
pprint(distinction(a_text))

# --- without Counter --- 

def distinction_analog(string):
    special_char = '@_!#$%^&*()<>?/\|}{~:;[]'
    for i in special_char: 
        string = string.replace(i, "")
    str_lst: list[str] = string.lower().split(" ")
    analog = {}
    for word in str_lst: 
        analog[word] = str_lst.count(word)
    return analog

pprint(distinction_analog(a_text))
```
